```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("clean_team_stats.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
#df = df.dropna()
df.head()
```

```python
df.dtypes
```

```python
df=df.drop(['team_no','variable', 'total_penalties_yards'], axis=1)
```

```python
df.dtypes
```

```python
FIELDS_TO_DROP = ['first_downs', 'third_down_percentage', 'yards_per_rush_attempt', 'rushing_touchdowns',
                 'passing_touchdowns']
```

```python
X = df.drop(FIELDS_TO_DROP, 1).dropna().drop_duplicates()
y = df[["first_downs","third_down_percentage"]].values
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
parameters = {'bootstrap': False,
              'min_samples_leaf': 3,
              'n_estimators': 50,
              'min_samples_split': 10,
              'max_features': 'sqrt',
              'max_depth': 6}
model = RandomForestRegressor(**parameters)
model.fit(X_train, y_train)
print(model.predict(X_test).astype(int), y_test)
```
