```python
# Dependencies
import pandas as pd
```

```python
file = "team/total_team_stats.csv"

df = pd.read_csv(file)
df.head()
```

```python
#replace spaces with underscores
df.columns = df.columns.str.replace(' ', '_')
df.head()
```

```python
#lowercase all column headers
df.columns = map(str.lower, df.columns)
df.head()
```

```python
#drop some bad data columns
drop = df.drop(columns=['3rd_down_efficiency', '4th_down_efficiency', 'touchback_percentage', 'possession_time_seconds', 'turnover_ratio'])
drop.head()
```

```python
#rename some columns
renamed = drop.rename(columns={"1st_downs":"first_downs", 
                               "1st_downs_by_penalty":"first_downs_by_penalty", 
                               "3rd_down_%":"third_down_percentage", 
                               "4th_down_%":"fourth_down_percentage", 
                               "comp-att":"comp_att", 
                               "fg:_good-attempts":"fg_good_attempt", 
                               "fumbles-lost":"fumbles_lost", 
                               "int:_total":"int_total", 
                               "kickoffs:_total":"kickoffs_total", 
                               "passing_1st_downs":"passing_first_downs",
                               "punt:_total":"punt_total",
                               "punt:_total_yards":"punt_total_yards",
                               "rushing_1st_downs":"rushing_first_downs",
                               "sacks-yards_lost":"sacks_yards_lost",
                               "total-yards":"total_penalties_yards",
                               "total_yards":"total_offensive_yards"
                              })
renamed.head()
```

```python
#list all renamed columns before splitting up data in certain columns
renamed.columns.tolist()
```

```python
#start splitting some data into two columns
renamed[['completed_passes','attempted_passes']] = renamed.comp_att.str.split("-",expand=True)
renamed.head()
```

```python
renamed[['field_goals_completed','field_goals_attempted']] = renamed.fg_good_attempt.str.split("-",expand=True)
renamed.head()
```

```python
renamed[['total_fumbles','fumbles_lost']] = renamed["fumbles_lost"].str.replace(':', '-').str.split("-", expand=True)
renamed.head()
```

```python
renamed[['defensive_interception','yards_after_interception']] = renamed["int_total"].str.split("-", expand=True)
renamed.head()
```

```python
renamed[['total_kickoffs_received','yards_off_kickoff_received']] = renamed["kickoffs_total"].str.split("-", expand=True)
renamed.head()
```

```python
renamed[['total_punts_received','yards_off_punts_received']] = renamed["punt_total"].str.split("-", expand=True)
renamed.head()
```

```python
renamed[['total_punts_kicked','total_punt_yards']] = renamed["punt_total_yards"].str.split("-", expand=True)
renamed.head()
```

```python
renamed[['total_defensive_sacks','yards_lost_from_sacks']] = renamed["sacks_yards_lost"].str.split("--", expand=True)
renamed.head()
```

```python
renamed[['total_penalties','total_yards_penalized']] = renamed["total_penalties_yards"].str.split("-", expand=True)
renamed.head()
```

```python
#list all columns
renamed.columns.tolist()
```

```python
#drop old columns
team_stats = renamed.drop(columns=['comp_att', 'fg_good_attempt', 'fumbles_lost', 'int_total', 'kickoffs_total', 'punt_total','punt_total_yards','sacks_yards_lost','total_penalties_yards'])
```

```python
#confirm new columns
team_stats.columns.tolist()
```

```python
#new clean team_stats to csv
team_stats.to_csv("team/cleaned_team_stats.csv")
```
