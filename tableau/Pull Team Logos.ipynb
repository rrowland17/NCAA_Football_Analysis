```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
image_list = []
parent_div = soup.findAll("div", {"class": "Wrapper bg-clr-white br-5 mb3 pa5"})
for tag in parent_div:
    images = tag.findAll("img")
    for image in images:
#         images.append(image)
       image_list.append({'image_url': image['src'], 'team_name': image['title']})
```
